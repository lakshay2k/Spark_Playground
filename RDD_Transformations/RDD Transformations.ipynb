# Transformations

### Remove Data File form Databricks File System

```
dbutils.fs.rm("dbfs:/FileStore/tables/data.txt")
```

### FlatMap

```
data = sc.textFile("dbfs:/FileStore/tables/mydata.txt")
rdd = data.flatMap(lambda x : x.split(" "))
rdd.collect()
```

### Filter

```
list1 = [1,2,3,4,5]
data = sc.parallelize(list1)
f=data.filter(lambda x : x!=5)
f.collect()
```

### Distinct

```
data = sc.textFile("dbfs:/FileStore/tables/mydata.txt")
rdd = data.flatMap(lambda x : x.split(" ")).distinct()
rdd.count()
```

### MapValues

```
rdd =  sc.parallelize([(1,5),(2,8),(3,10),(4,56),(5,65),(6,55)])
rdd.mapValues(lambda x : x+1).collect()
```

### FlatMapValues

```
data = sc.textFile("dbfs:/FileStore/tables/user_address.txt")
a = data.map(lambda x: x.split("\t"))
b = a.map(lambda x : (x[0],x[1]))
c = b.flatMapValues(lambda x:x.split(";"))
c.collect()
```

### Key

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt")
a = data.flatMap(lambda x : x.split(" "))
b = a.map(lambda x : (x,1)).keys()
b.collect()
```

### Values

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt")
a = data.flatMap(lambda x : x.split(" "))
b = a.map(lambda x : (x,1)).values()
b.collect()
```

### sortByKey

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt")
a = data.flatMap(lambda x : x.split(" "))
b = a.map(lambda x : (x,1)).sortByKey()
b.collect()
```

### groupByKey

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt")
a = data.flatMap(lambda x : x.split(" "))
b = a.map(lambda x : (x,1)).groupByKey()
b.collect()
```

### reduceByKey

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt")
a = data.flatMap(lambda x : x.split(" "))
b = a.map(lambda x : (x,1)).reduceByKey(lambda v1,v2: v1+v2)
b.collect()
```

### Join on RDDS
Perform an inner join between two RDDs

```
data = sc.textFile("dbfs:/FileStore/tables/spark.txt").flatMap(lambda x : x.split(" ")).map(lambda x:(x,1))
data2 = sc.textFile("dbfs:/FileStore/tables/hadoop.txt").flatMap(lambda x:x.split()).map(lambda x:(x,1))
data3 = data.join(data2)
data3.collect()
```

### Join on files

```
data = sc.textFile("dbfs:/FileStore/tables/log.txt").map(lambda x : x.split(" ")).map(lambda k:(k[9],k[0]))
data2 = sc.textFile("dbfs:/FileStore/tables/userdetails.csv").map(lambda x:x.split(",")).map(lambda k:(k[0],k[1]))
data3 = data.join(data2)
data3.collect()
```

### Union
Perform a union between two RDDs

```
a=sc.textFile("dbfs:/FileStore/tables/city1.txt")
b=sc.textFile("dbfs:/FileStore/tables/city2.txt")
rdd = a.union(b)
rdd.collect()
```

### Zip
Return key-value pairs with the first element in each RDD to second element in other RDD

```
a=sc.textFile("dbfs:/FileStore/tables/city1.txt")
b=sc.textFile("dbfs:/FileStore/tables/city2.txt")
rdd = a.zip(b)
rdd.collect()
```

### Subtract
Creates new RDD from first RDD having elements which are not in second RDD

```
a=sc.textFile("dbfs:/FileStore/tables/city1.txt")
b=sc.textFile("dbfs:/FileStore/tables/city2.txt")
rdd = a.subtract(b)
rdd.collect()
```

### Intersection
Return the intersection of elements between two RDDs

```
a=sc.textFile("dbfs:/FileStore/tables/city1.txt")
b=sc.textFile("dbfs:/FileStore/tables/city2.txt")
rdd = a.intersection(b)
rdd.collect()
```

### mapPartitions(func)
Similar to map, but runs separately on each partition (block) of the RDD.

### CoGroup
When called on datasets of type (K, V) and (K, W), returns a dataset of (K, (Iterable<V>, Iterable<W>)) tuples. This operation is also called groupWith.

```
rdd1 = sc.parallelize([("A",1),("B",2),("C",3)])
rdd2 = sc.parallelize([("B",4),("D",5)])
cg = rdd1.cogroup(rdd2)
cg.collect()
```
