## Word Count Problem

### Step 1: Read the file spark.txt

```python
myfile = sc.textFile("FileStore/tables/spark.txt")
```

### Step 2: Split the file

```python
myfile_split = myfile.flatMap(lambda x : x.split(" "))
```

### Step 3: Make a pair RDD by adding 1 to each word

```python
pair_rdd = myfile_split.map(lambda x: (x,1))
```

### Step 4: word count

```python
count = pair_rdd.reduceByKey(lambda a,b : a+b)
```

### Step 5: Save the result into file

```python
count.saveAsTextFile("FileStore/tables/myresult.txt")
```

### Step 6: Verify the Result

```python
verify = sc.textFile("FileStore/tables/myresult.txt")
verify.collect()
```

## Caching

### Step 1: Create a RDD using parallelize()

```python
rdd = sc.parallelize([1,6,8,63,8,1,2,3,5])
```

### Step 2: Cache the RDD

```python
rdd.cache()
```

### Step 3: Check if RDD is cached or not

```python
rdd.is_cached
```

### Step 4: Remove persistance using unpersist()

```python
rdd.unpersist()
```

### Step 5: Check if RDD is cached or not

```python
rdd.is_cached
```

### Step 6: Cache the RDD using persist()

```python
rdd.persist()
```

### Step 7: Check if RDD is cached or not

```python
rdd.is_cached
```

### Step 8: Remove persistance using unpersist()

```python
rdd.unpersist()
```

### Step 9: Check if RDD is cached or not

```python
rdd.is_cached
```
