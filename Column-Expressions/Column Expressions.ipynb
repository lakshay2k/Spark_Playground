### Step1 : Import **``student_marks_data.csv``** into Databricks tables folder

### Step2 :  Import SQLContext

```python
from pyspark.sql import SQLContext
```

### Step3 :  Load the  and creating DataFrame and display 5 elements
Here, we are inferring the schema from the first row of the data-containing document

```python
fileName = "dbfs:/FileStore/tables/student_marks_data.csv"
studentMarksDF = spark.read.csv(fileName,header = "true", inferSchema = "true")
studentMarksDF.show(5)
```

### Step4 : Print the schema using printSchema

```python
studentMarksDF.printSchema()
```

### Step5 :  Adding a New Coloumn
The following statement applies exponential function on test1 coloumn values and generates a new coloumn expTest1

```python
#to perform exp() operation on existing coloumn
from pyspark.sql.functions import exp

#returns a new dataframe with an additional new coloumn with name "expTest1"
df_with_exp = studentMarksDF.withColumn("expTest1", exp("test1"))
#displays first 2 records of new dataframe
df_with_exp.show(2)
```

### Step6 :  Generating Random values in the new Coloumn

```python
#to call rand() function to generate random values
from pyspark.sql.functions import rand

#returns a new dataframe with additional column "random" having random values
df_with_rand = studentMarksDF.withColumn("rand()",rand())

#displays first 2 records of new dataframe
df_with_rand.show(2)
```

### Step7 :  Calculating Percentage of marks for every student

```python
#giving an alias to the dataframe studentMarksDF
df = studentMarksDF
#adds the new column having percentage marks values of every student in each subject
df_with_percentage = df.withColumn("Percentage",(df.test1+df.test2+df.test3+df.test4)/4)
#displays all the records of dataframe
df_with_percentage.show(5)
```

### Step8 :  Use of "when" statement and renaming the Column
"when" & "othewise" statement is use to specify some condition and perform a desired action if that condition holds or doesnn't holds

```python
#functions is required to use when statement so we import it from pyspark.sql
from pyspark.sql import functions as F
k = df_with_percentage.select(df.Student_id,F.when(df_with_percentage.Percentage>50,"Pass").otherwise("Fail").alias("Result"))
#ruturns a subset dataframe based on a condition
k.show(10)
```

### Step9:  Sorting a column
orderBy() sorts the dataframe based on the values of a column

```python
df_with_percentage.orderBy('Percentage').show(5)
```

### Step10 :  Sort using SQL

```python
#registering 'percentage_table' as temptable to use sql queries
df_with_percentage.createOrReplaceTempView('percentage_table')
```

### Step11 :  Execute sql query 'select * from percentage_table ORDER BY Percentage' and display the result

```python
spark.sql("select * from percentage_table ORDER BY Percentage").show()
```

### Step12 :  Droping a column
drop() function is used to remove a column 'year_ID'

### Step14 : Replacing contents of the column and show 10 elements

```python
#replacement of Subject_type coloumn values 'PRC', 'Practical' to 'THE', 'Theory'
from pyspark.sql.functions import *
newDF = studentMarksDF.withColumn("Subject_type",regexp_replace("Subject_type","PRC","Practical")).withColumn("Subject_type",regexp_replace("Subject_type","THE","Theory"))
newDF.show(10)
```
